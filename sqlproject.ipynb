```python
# Defines a function called select to execute a SQL select query and returns the result as a DataFrame.
def select(q):
    # Calls the previously defined create_connection function to establish a connection to the MySQL database.
    connection = create_connection()
    
    # Checks if a valid connection object was returned.
    if connection:
        # Starts a try-except block to handle potential exceptions when executing the select query.
        try:
            # Executes the provided SQL select query on the established connection and returns the result as a DataFrame.
            df = read_sql_query(q, connection)
            
            # Returns the resulting DataFrame.
            return df
        # Catches any exceptions of type 'Error' that might occur during the query execution.
        except Error as e:
            # Prints the error message to the console.
            print("Error executing the query:", e)
        # The finally block ensures that the following code is executed whether or not an exception occurred.
        finally:
            # Closes the connection to the MySQL database.
            connection.close()
    # Returns None if the connection wasn't established or if there was an error in query execution.
    return None
```

### Since the data is in wide formate that contains four awareness score, we sum them into an aggragte variable. Then we order it by awareness score

```python
# Executes a SQL select query using the select function to compute the awareness_score for each participant.
awarness_score = select("""
    # Selects the participant_id column and computes the sum of awareness columns,
    # and labels this computed column as 'awareness_score'.
    SELECT participant_id,
           awareness1 + awareness2 + awareness3 + awareness4 as awareness_score
    # Specifies the source table 't1_cleaned' from which the data is to be fetched.
    FROM t1_cleaned
    # Orders the result set based on the computed 'awareness_score' in ascending order by default.
    ORDER BY awareness_score
""")

```

### We can also use subquries to check those who have above average awareness score

```python
# Executes a SQL select query using the select function to fetch participant IDs with above-average awareness scores.
awarness_score_above_average = select("""
    # Selects the participant_id column and computes the sum of awareness columns,
    # and labels this computed column as 'awareness_score'.
    SELECT participant_id,
           awareness1 + awareness2 + awareness3 + awareness4 as awareness_score
    # Specifies the source table 't1_cleaned' from which the data is to be fetched.
    FROM t1_cleaned
    # Applies a WHERE clause to filter results where the computed 'awareness_score' for a participant exceeds the average score of all participants.
    WHERE awareness1 + awareness2 + awareness3 + awareness4 > 
        (SELECT AVG(awareness1 + awareness2 + awareness3 + awareness4)
         FROM t1_cleaned)
    # Orders the result set based on the computed 'awareness_score' in ascending order by default.
    ORDER BY awareness_score
""")
```

### We can also use join function to join demographic tables on t1, so it can show participant job role

```python
# Executes a SQL select query using the select function to fetch participant IDs, their awareness scores, 
# and a demographic detail by joining the t1_cleaned and demographic tables.
awarness_score_job = select("""
    # Selects the participant_id column from the t1_cleaned table (aliased as t1), computes the sum of awareness1, awareness2, 
    # awareness3, and awareness4 columns, and labels this computed column as 'awareness_score'. Also selects the C2 column from 
    # the demographic table (aliased as d).
    SELECT participant_id,
           awareness1 + awareness2 + awareness3 + awareness4 as awareness_score,
           d.C2
    # Specifies the source table 't1_cleaned' and aliases it as 't1'.
    FROM t1_cleaned as t1
    # Performs an inner join with the 'demographic' table (aliased as 'd') based on the participant IDs in both tables.
    JOIN demographic as d
    ON t1.participant_id = d.C1
    # Orders the result set based on the computed 'awareness_score' in ascending order by default.
    ORDER BY awareness_score
""")

```

### We use both t1 and t2 tables, where t1 means before app use and t2 means after app use. We use subqureries to select awareness score in both t1 and t2 table, and find those who have improved their awareness score

```python
# Executes a SQL select query using the select function to fetch participants with an improvement in awareness scores 
# from the t1_cleaned table to the t2_cleaned table.
awarness_score_improve = select("""
    # Selects participant_id, awareness scores from t1_cleaned (labeled as t1_awareness_score) 
    # and awareness scores from t2_cleaned (labeled as t2_awareness_score).
    SELECT participant_id,
           t1_awareness_score,
           t2_awareness_score
    FROM 
        # Constructs a derived table using a subquery to compute awareness scores for each participant in both tables.
        (SELECT participant_id,
                t1.awareness1 + t1.awareness2 + t1.awareness3 + t1.awareness4 as t1_awareness_score,
                t2.awareness1 + t2.awareness2 + t2.awareness3 + t2.awareness4 as t2_awareness_score
         FROM t1_cleaned as t1
         # Performs an inner join with the 't2_cleaned' table using the 'participant_id' as the common key.
         JOIN t2_cleaned as t2 USING(participant_id)
        ) as derived
    # Filters results to only include participants whose awareness scores have increased from t1_cleaned to t2_cleaned.
    WHERE t1_awareness_score < t2_awareness_score
""")

```
